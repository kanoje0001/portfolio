# Que 1: Swapping first and last element of list

```python
ls = [1,2,3,4,5,11,12,32]
print("List before swapping: ",ls)
ls[0],ls[-1] = ls[-1],ls[0]
print("List after swapping: ",ls)
```

# Que 2: Length of list using for loop

```python
length = 0;
for i in ls:
    length += 1
print("Length of list: ", length)
```

# Que 1: Multiplication of matrix with identity matrix

```python
from numpy import *

a = [[9,10,11],
    [12,13,14],
    [15,16,17]]

i=[[1,0,0],
  [0,1,0],
  [0,0,1]]

res = dot(a,i)

print("Matrix a after multiplying with an identity matrix: "res)
```

# Que 2: Write a program to Check whether a NumPy array contains a specified row

```python
ls = [[1,2,3],[11,24,11],[23,53,55],[1323,3223,1342]]

arr = [23,53,55]
not_present = [1,4,5]

print(arr,"is present in ",ls," = ",arr in ls)
print(not_present,"is present in",ls," = ",not_present in ls)
```
